```
#Numerical Distributions and data types for demog
demog = pd.read_csv(file_path)
print(demog.describe())
demog.dtypes
```

```
#Missing data - no major concern
print(demog.isnull().any())
demog.isnull().sum()/demog.shape[0]
```

```
#Check usual values in non-numeric columns
for col in demog.columns:  # iterate over all columns
  if demog[col].dtype not in [np.int64, np.float64]:  # ignore integer and float columns
    # print a header, all unique values, and a separator
    print('Unique Values for {}'.format(col))  
    print(demog[col].unique())   
    print('')
```

```
#Data cleaning on State and County to improve matching
demog["State"] = demog["State"].str.upper()
demog["County"] = demog["County"].str.upper()
demog["County"] = demog["County"].str.replace(' CENSUS AREA', '', regex=True)
demog["County"] = demog["County"].str.replace(' CITY AND BOROUGH', '', regex=True)
demog["County"] = demog["County"].str.replace(' COUNTY', '', regex=True)
demog["County"] = demog["County"].str.replace(' BOROUGH', '', regex=True)
demog["County"] = demog["County"].str.replace(' MUNICIPALITY', '', regex=True)
demog["County"] = demog["County"].str.replace(' MUNICIPIO', '', regex=True)
demog["County"] = demog["County"].str.replace(' PARISH', '', regex=True)
demog["County"] = demog["County"].str.replace('-', ' ', regex=True)
demog["County"] = demog["County"].str.replace('.', '', regex=True)
demog["County"] = demog["County"].str.replace('\'', '', regex=True)
demog.loc[demog["CountyId"] == 72119, "County"] = 'RIO GRANDE'
demog.loc[demog["CountyId"] == 72075, "County"] = 'JUANA DIAZ'
demog["County"][demog["CountyId"]==35013] = "DONA ANA"
demog["County"][demog["CountyId"]==51520] = "BRISTOL"
demog["County"][demog["CountyId"]==51750] = "RADFORD"
demog["County"][demog["CountyId"]==51775] = "SALEM"
```

```
#Check if State and County uniquely determined CountyId
count_state_county = pd.DataFrame(demog.groupby(['State','County'])['CountyId'].count())
print(count_state_county[count_state_county['CountyId']!=1])
```

```
county = demog[["CountyId","State",'County']]
```

```
#Numerical Distributions and data types for equip
equip = pd.read_csv(file_path2)
print(equip.head())
print(equip.describe())
equip.dtypes
```

```
#Missing data - no major concern
#Some County is missing so they won't be matched. We'll checking  % of counties matched later
print(equip.isnull().any())
equip.isnull().sum()/equip.shape[0]
```

```
#Check usual values in non-numeric columns
for col in equip.columns:  # iterate over all columns
  if equip[col].dtype not in [np.int64, np.float64]:  # ignore integer and float columns
    # print a header, all unique values, and a separator
    print('Unique Values for {}'.format(col))  
    print(equip[col].unique())   
    print('')
```

```
#Data cleaning for State and County to improve matching
equip["County"] = equip["County"].str.replace('SAINT', 'ST', regex=True)
equip["County"] = equip["County"].str.replace('-', ' ', regex=True)
equip["State"] = equip["State"].str.replace('DELAWRE', 'DELAWARE')
equip["County"] = equip["County"].str.replace('DE KALB', 'DEKALB')
equip["County"] = equip["County"].str.replace('LA PORTE', 'LAPORTE')
equip["County"][(equip["State"]=='FLORIDA') & (equip["County"]=='DE SOTO')] = 'DESOTO'
equip["County"][(equip["State"]=='ILLINOIS') & (equip["County"]=='LA SALLE')] = 'LASALLE'
equip["County"][(equip["State"]=='LOUISIANA') & (equip["County"]=='LA SALLE')] = 'LASALLE'

equip_m = equip.merge(county, how='left', 
                                      left_on=['State','County'], 
                                         right_on=['State','County'])
print(equip_m.head())
missing = equip_m[['State','County']][equip_m['CountyId'].isnull()]
print(missing["County"].unique()) 
print(missing)
print(missing.describe())  
print(missing.shape)
```

```
#% of counties matched - goal is above 98%
1-missing.shape[0] /equip_m.shape[0]
```

```
#Too many unique values for Item Name, so we will use key words match to determine if it's a weapon
equip_m["Item Name"].unique().shape
```

```
weaponList = ['RIFLE', 'PISTOL', 'SHOTGUN','EXPLOSIVE','ASSAULT','AMMUNITION','MACHINE GUN','GUNS','GRENADE','AUTOMATIC GUN','CARBINE','BAYONET']
weaponString = '|'.join(weaponList)
print(weaponString)
equip_m["Item Name"][equip_m["Item Name"].isna()] = ""
equip_m["Type"] = 'OTHER'
equip_m["Type"][equip_m["Item Name"].str.contains(weaponString)] = "WEAPON"
print(equip_m)
#COMBAT 
#SCOPE
```
